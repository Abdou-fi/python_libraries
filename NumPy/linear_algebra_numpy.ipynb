We start by importing the namespace of the numpy library under the alias np, and by defining an invertible matrix M.

```python
import numpy as np

# for better display of results, precision is set to 4 digits, 
# and scientific notation is suppressed.
np.set_printoptions(precision = 4, suppress=True)   

# we define the numpy matrix M
M = np.array( [[1, 2, 3],
               [2, 3, 4],
               [3, 4, 6]] )
```

The classical operations related to the resolution of linear systems are available in the linalg library included in numpy.

```python
print(np.linalg.det(M)) # determinant
```

```python
M_inv = np.linalg.inv(M) # inverse
print(M_inv)
```

We can check that minv is indeed the (numerical) inverse of M: $M^{-1}⋅M=I$, where $I$ is the identity matrix.
 

```python
print(np.dot(M_inv, M))
```

To solve the linear system  $M⋅X=B$, we use the function linalg.solve(). Note that the matrix must be square and invertible

This function is more efficient than computing the inverse of M and then multiplying it by B, because it uses a more efficient algorithm to solve the system directly.

$M⋅X=B$ is equivalent to the algebraic linear system:
$$\begin{cases}
x_1 + 2x_2 + 3x_3 = 2 \\
2x_1 + 3x_2 + 4x_3 = 2 \\
3x_1 + 4x_2 + 6x_3 = 3
\end{cases}$$

or the matrix form:
$$\begin{pmatrix}
1 & 2 & 3 \\
2 & 3 & 4 \\
3 & 4 & 6 
\end{pmatrix}  \begin{pmatrix}x_1 \\
x_2 \\
x_3 \end{pmatrix} = \begin{pmatrix}2 \\ 2 \\ 3 \end{pmatrix}$$

We can solve this system using numpy:

```python
B = np.array([2, 2, 3])
X = np.linalg.solve(M, B)
print(X)
```

```python
X=[-1, 0, 1]  # change from -0 to 0 to avoid Python approximation errors
print(np.dot(M, X) == B) # we check that the solution is correct
```
