# Data Cleaning techniques

Data cleaning is a crucial step in data analysis, ensuring the data is accurate, consistent, and ready for analysis. Pandas, a powerful Python library, provides various tools and functions to handle data cleaning tasks efficiently.

## I. Handling missing data

Missing data is acommon issue in datasets. Pandas provides several methods to handle missing values, such as dropping rows or columns, filling with a specific value, or using interpolation techniques.

```python
import pandas as pd
import numpy as np

# Sample dataset
data = {
    'Name': ['Ahmed', 'SARAH', 'Imad', 'KARIM'],
    'Age': [25, 30, np.nan, 40],  # missing value
    'Salary': [5000, 6000, 7000, np.nan]  # missing value
}
df = pd.DataFrame(data)
# drop rows with missing values
df.dropna(inplace=True)
# fill missing values with a specific value
df.fillna(0, inplace=True)
# Fill missing values with the mean of the column
df['Age'] = df['Age'].fillna(df['Age'].mean())
# fill missing values using interpolation
df.interpolate(method='linear', inplace=True)
```

## II. Removing duplicates

Duplicate data can lead to inaccurate analysis and skewed results. Pandas offers methods to identify and remove duplicate rows or columns from a DataFrame.

```python
# Remove duplicate rows 
df.drop_duplicates( inplace=True)
# Remove duplicate rows based on specific columns 
df.drop_duplicates(subset=['Name', 'Age'], inplace=True)
```

## III. Data transformation
Data transformation involves converting data into a more suitable format for analysis. Pandas provides functions to perform operations like data type conversion, string manipulation, and date/time handling.

```python
import pandas as pd

# Sample dataset
data = {
    'Name': ['Ahmed', 'SARAH', 'Imad', 'KARIM'],
    'Age': ['25', '30', 35, '40.5'],  # stored as strings initially
    'Join_Date': ['2021-05-21', '2020/06/15', '15-07-2019', 'April 10, 2018']  # various date formats
}
df1 = pd.DataFrame(data)
print(df1)
print("\n =============================================")

df1['Age'] = df1['Age'].astype('int')
df1['Name'] = df1['Name'].str.capitalize()
df1['Join_Date'] = pd.to_datetime(df1['Join_Date'], format='mixed')
df1['Age'] = pd.to_numeric(df1['Age'], errors='coerce')

print(df1)
```

## V. Data filtering
Data filtering is the process of selecting a subset of data based on specific criteria. Pandas provides powerful filtering capabilities using boolean indexing and conditional statements.

```python
import pandas as pd
# Random wage dataset
data = {
    'Person': ['Alex', 'Bella', 'Chris', 'Dana', 'Eli', 'Faye', 'Gabe', 'Hana', 'Ivan', 'Julia'],
    'Gender': ['Man', 'Woman', 'Man', 'Woman', 'Man', 'Woman', 'Man', 'Woman', 'Man', 'Woman'],
    'Age': [28, 35, 51, 31, 27, 39, 45, 33, 52, 36],
    'Job_Role': ['Data Analyst', 'Graphic Designer', 'Software Engineer', 'Data Analyst',
                 'Data Analyst', 'Graphic Designer', 'Data Analyst', 'Software Engineer',
                 'Software Engineer', 'Data Analyst'],
    'Wage': [3200, 2900, 4500, 3100, 3700, 2800, '4100', 3300, 3900, 3000],
    'Join_Date': ['2022-01-10', '2021-12-15', '2020-06-20', '2019-11-01', 
                  '2021-03-25', '2020-08-14', '2018-04-10', '2017-09-19',
                  '2022-07-30', '2019-01-05']
}

df2 = pd.DataFrame(data)
df2['Wage'] = df2['Wage'].astype(int)
df2['Age'] = df2['Age'].astype(int)
df2['Join_Date'] = pd.to_datetime(df2['Join_Date'])
print(df2)
print("\n==========================================")
df_sorted=df2.sort_values(by='Wage', ascending=False, na_position='first')
print(df_sorted)

print("\n==========================================")
# Filter rows based on a condition 
filtered_df = df2[df2['Age'] > 50] 

# Filter rows based on multiple conditions 
filtered_df = df2[(df2['Age'] > 50) & (df2['Gender'] == 'Man')]
print(filtered_df)

print("\n==========================================")
df_grouped = df2.groupby('Gender')['Wage'].min()
print(df_grouped)

```

## VII. Data merging and joining
Merging and joining data from multiple sources is a common task in data analysis. Pandas provides methods to combine datasets based on common columns or indexes.

```python
import pandas as pd

# First DataFrame: Basic employee info
df1 = pd.DataFrame({
    'Employee_ID': [101, 102, 103, 104],
    'Name': ['Amir', 'Sarah', 'Karim', 'Yazid'],
    'Department': ['Sales', 'Marketing', 'HR', 'Finance']
})

# Second DataFrame: Performance review scores
df2 = pd.DataFrame({
    'Employee_ID': [101, 102, 103, 104],
    'Review_Score': [88, 92, 79, 85]
})

# Merge on the common column 'Employee_ID'
merged_df = pd.merge(df1, df2, on='Employee_ID')
print(df1)
print("\n==========================================")
print(df2)
print("\n==========================================")
print(merged_df)


# Join two DataFrames based on indexes 

# First DataFrame: Employee Info
df3 = pd.DataFrame({
    'Name': ['Amir', 'Bianca', 'Carlos', 'Dina'],
    'Department': ['Sales', 'Marketing', 'HR', 'Finance']
}, index=[101, 102, 103, 104])  # Employee_ID as index

# Second DataFrame: Performance Scores
df4 = pd.DataFrame({
    'Review_Score': [88, 92, 79, 85]
}, index=[101, 102, 103, 104])  # Matching index

# Join on index
joined_df = df3.join(df4, how='inner')
# Join two DataFrames based on indexes 
joined_df = df3.join(df4, how='inner' )
print(df3)
print("\n==========================================")
print(df4)
print("\n==========================================")
print(joined_df)
```

## VIII. Data reshaping
Data reshaping involves transforming the structure of a DataFrame, such as pivoting or unpivoting data. Pandas offers functions like melt and pivot to reshape data for better analysis.

```python
# Unpivot (melt) data 
import pandas as pd

# Wide-format data
df5 = pd.DataFrame({
    'Student_ID': ['001', '002', '003'],
    'Name': ['Ali', 'Aymen', 'Mona'],
    'Math': [85, 78, 93],
    'Science': [90, 82, 88],
    'History': [75, 85, 80]
})

melted_df = pd.melt(
    df5,
    id_vars=['Student_ID', 'Name'],  # columns to keep
    var_name='Subject',
    value_name='Score'
)
print(df5 )
print("\n==========================================")
print(melted_df )


# Pivot data 

# Long-format data (after melt)
df6 = pd.DataFrame({
    'Student_ID': ['001', '001', '001', '002', '002', '002', '003', '003', '003'],
    'Name': ['Ali', 'Ali', 'Ali', 'Brenda', 'Brenda', 'Brenda', 'Carlos', 'Carlos', 'Carlos'],
    'Subject': ['Math', 'Science', 'History', 'Math', 'Science', 'History', 'Math', 'Science', 'History'],
    'Score': [85, 90, 75, 78, 82, 85, 93, 88, 80]
})

print("\n==========================================")
print(df6)
pivoted_df = df6.pivot(index='Student_ID', columns='Subject', values='Score')
print("\n==========================================")
print(pivoted_df)
```

## XII. Data validation
Data validation is the process of ensuring that data adheres to specific rules, constraints, or formats. Pandas provides methods to validate data and handle violations, such as raising errors or applying custom functions.

```python
data = {
    'Person': ['Alex', 'Bella', 'Chris', 'Dana', 'Eli', 'Faye', 'Gabe', 'Hana', 'Ivan', 'Julia'],
    'Gender': ['Man', 'Woman', 'Man', 'Woman', 'Man', 'Woman', 'Man', 'Woman', 'Man', 'Woman'],
      'Age': [28, 35, '12', 31.6, 27, 39, 45, 33, 52, 75],
    'Job_Role': ['Data Analyst', 'Graphic Designer', 'Software Engineer', 'Data Analyst',
                 'Data Analyst', 'Graphic Designer', 'Data Analyst', 'Software Engineer',
                 'Software Engineer', 'Data Analyst'],
    'Wage': [3200.9, 2900.8, 4500.8, 3100.4, 3700.1, 2800.5, '4100.0', 3300.2, 3900.3, 3000.9],
    'Join_Date': ['2022-01-10', '2021-12-15', '2020-06-20', '2019-11-01', 
                  '2021-03-25', '2020-08-14', '2018-04-10', '2017-09-19',
                  '2022-07-30', '2019-01-05']
}

df8 = pd.DataFrame(data)

# Validate data types 
df8 = df8.astype({'Age': 'int', 'Wage': 'float'}) 
print(df8)
print("\n =================================================")

# Apply custom validation function 
def validate_age(age): 
    if age < 18 or age > 70: 
        # raise ValueError('Invalid age' ) 
        print(f"Invalid age found: {age}")
        return ""    
    else :
        return age 
df8['Age'] = df8['Age'].apply(validate_age)
print(df8)

# df8['Valid_Age'] = df8['Age'].between(18, 70)
# print(df8['Valid_Age'])
```
